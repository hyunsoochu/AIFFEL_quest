```python

import re

with open('/content/drive/MyDrive/Colab Notebooks/Avengers.txt','r') as f:
    text = f.read()

def processingtext(text):
  lower_text = text.lower()
  cleaned_text = re.sub(r'[^a-zA-Z0-9\s]', '', lower_text)
  split_text = cleaned_text.split()

  return split_text

```

```python
dic = {}

split_text = processingtext(text)

for i in split_text:
  if i not in dic:
    dic[i] = 1
  else:
    dic[i] +=1
print(dic)
```

```python
sort_dic = sorted(dic.items(), key = lambda item:item[1], reverse=True)

vocab = {}
i = 0

for item in sort_dic:
  vocab[item[0]] = i
  i += 1

print(vocab)
```

```python
content = str(input("텍스트를 입력해주세요 : "))

intList = []

def changeText(text):
  split_content = processingtext(content)
  for i in split_content:
    intList.append(vocab[i])
  return intList

changeText(content)
```
